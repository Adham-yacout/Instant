```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
income=pd.read_csv('canada_per_capita_income.csv')
```

```python
income
```

```python
%matplotlib inline
income = income.rename(columns={'per capita income (US$)': 'incomerate'})
plt.xlabel('year')
plt.ylabel('INCOME')
plt.scatter(income.year,income.incomerate,color='green',marker='+')
```

```python
df=income.drop("incomerate",axis='columns')
df
```

```python
incomerate=income.incomerate
incomerate
```

```python
reg = linear_model.LinearRegression()
reg.fit(df,incomerate)
reg.score(df,incomerate)
```
